<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkgreen;">
        <strong>
        <h5 style="color: black;"><br>
*** To solve this, we will follow these steps:<br>

1. Convert the "Olympic Date" column to a datetime object using pd.to_datetime().<br>
2. Use resample() to group the data by year and sum the wins per year.<br>
3. Filter the years where the total wins are greater than 0.<br>
4. As a bonus, replace all 0 values with NaN and use dropna() to remove rows with NaN values.
</div>

```python
# Step 1: Convert the Olympic Date to datetime format
df['Olympic Date'] = pd.to_datetime(df['Olympic Date'], format='%d/%m/%Y');

# Step 2: Set the Olympic Date as the index for resampling
df.set_index('Olympic Date', inplace=True);

# Step 3: Resample the data by year and sum the 'Wins' for each year
yearly_wins = df.resample('Y')['Wins'].sum();

# Step 4: Filter the years with at least one win (greater than 0)
yearly_wins_filtered = yearly_wins[yearly_wins > 0]

# Bonus: Replace all zeros in the wins column with NaN and drop rows with NaN values
df['Wins'] = df['Wins'].replace(0, pd.NA);  # Replace 0 with NaN
df_cleaned = df.dropna(subset=['Wins']);  # Drop rows where 'Wins' is NaN

# Display the results
print("Yearly wins (filtered by years with at least one win):");
print(yearly_wins_filtered);

print("\nCleaned DataFrame with 0s replaced by NaN and dropped:");
print(df_cleaned);
```
